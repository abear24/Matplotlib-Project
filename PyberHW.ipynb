```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# File to Load (Remember to change these)
city_data = "data/city_data.csv"
ride_data= "data/ride_data.csv"

# Read the City and Ride Data
city_data="data/city_data.csv"
ride_data="data/ride_data.csv"

city_data_df=pd.read_csv(city_data)
ride_data_df=pd.read_csv(ride_data)

# Combine the data into a single dataset

merged_data=pd.merge(city_data_df,ride_data_df, on="city")
merged_data.head(50)

# Display the data table for preview
```

```python
#find average fare

average_fare=merged_data.groupby("city")["fare"].mean()
average_fare.head(50)

```

```python
#find total number of rides

total_rides=merged_data.groupby("city")["ride_id"].count()
total_rides.head(50)
```

```python
#find total drivers
total_drivers=merged_data.groupby("city")["driver_count"].mean()
total_drivers.head()
```

```python
#find city types

city_type =city_data_df.set_index("city")["type"]
city_type.value_counts()
```

```python
#clean variables and create new dataframe
pyber_data=pd.DataFrame({"Average Fare/City":average_fare,"Total Rides":total_rides,"Total Drivers":total_drivers,"City Type":city_type})
pyber_data.head(50)

```

```python
#retrieve variables and change variables for city types

urban=pyber_data.loc[pyber_data["City Type"]== "Urban"]
suburban=pyber_data.loc[pyber_data["City Type"]=="Suburban"]
rural=pyber_data.loc[pyber_data["City Type"]=="Rural"]


```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban=pyber_data.loc[pyber_data["City Type"]== "Urban"]
suburban=pyber_data.loc[pyber_data["City Type"]=="Suburban"]
rural=pyber_data.loc[pyber_data["City Type"]=="Rural"]

# Build the scatter plots for each city types
urban_plot=plt.scatter(urban["Total Rides"],urban["Average Fare/City"], color="coral", edgecolors="black",
s=urban["Total Drivers"]*20, label="Urban", alpha=0.5, linewidths=1.5)

suburban_plot=plt.scatter(suburban["Total Rides"],suburban["Average Fare/City"], color="lightskyblue", edgecolors="black",
s=suburban["Total Drivers"]*20, label="Suburban", alpha=0.5, linewidths=1.5)

rural_plot=plt.scatter(rural["Total Rides"],rural["Average Fare/City"], color="gold", edgecolors="black",
s=rural["Total Drivers"]*20, label="Rural", alpha=0.5,linewidths=1.5)

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data(2016)")
plt.xlabel("Total Number of Rides(Per City)")
plt.ylabel("Average Fare($)")

plt.grid(True)
# Create a legend
plt.legend(title ="City Types")
#loc="upper right")

# Incorporate a text label regarding circle size
plt.text(40,50, "Note: Circle size correlates with driver count per city.")

# Save Figure
plt.savefig("scatterplot.png")
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
city_pie=merged_data.groupby(["type"])["fare"].sum()
city_pie
# Build Pie Chart
city_pie.index
colors=["gold","lightskyblue","lightcoral"]
explode=(0,0,0.1)
plt.pie(city_pie,labels=city_pie.index, autopct="%1.2f%%", colors=colors, explode=explode,
        shadow=True, startangle=130)


plt.title("% of Total Fares by City Type")

# Save Figure/Show Figure
plt.savefig("piechart.png")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
city_pie2=merged_data.groupby(["type"])["ride_id"].count()
city_pie2
# Build Pie Chart
city_pie2.index
colors=["gold","lightskyblue","lightcoral"]
explode=(0,0,0.1)
plt.pie(city_pie2,labels=city_pie2.index, autopct="%1.2f%%", colors=colors, explode=explode,
        shadow=True, startangle=150)


plt.title("% of Total Rides by City Type")

# Save Figure/Show Figure
plt.savefig("piechart2.png")
plt.show()
```
